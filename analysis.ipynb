```python
# Encode categorical features
df_encoded = df_model.copy()

from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
import pandas as pd
```

### Binary Encoding

```python
# Binary categorical variables: default, housing, loan, y 
binary_cols = ['default', 'housing', 'loan', 'y']

# Create a label encoder
le = LabelEncoder()

# Apply label encoding to binary columns
for col in binary_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
    print(f"{col} mapping: {dict(zip(le.classes_, le.transform(le.classes_)))}")
```

### Ordinal Encoding

```python
# Education has a natural order
education_order = ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree', 'unknown']

# Check if our ordering contains all unique values
print(f"Education unique values: {df_encoded['education'].unique()}")

# Create an ordinal encoder for education
oe = OrdinalEncoder(categories=[education_order])
df_encoded['education_encoded'] = oe.fit_transform(df_encoded[['education']])

# Display the mapping
educ_mapping = {category: idx for idx, category in enumerate(education_order)}
print("Education level mapping:")
print(educ_mapping)
```

### One-Hot Encoding

```python
# Nominal categories: job, marital, contact, poutcome
nominal_cols = ['job', 'marital', 'contact', 'poutcome']

# Create and fit the one-hot encoder
ohe = OneHotEncoder(sparse_output=False, drop='first')

# Apply one-hot encoding
encoded_nominal = ohe.fit_transform(df_encoded[nominal_cols])

# Create a DataFrame with the one-hot encoded columns
encoded_nominal_df = pd.DataFrame(
    encoded_nominal, 
    columns=ohe.get_feature_names_out(nominal_cols),
    index=df_encoded.index
)

# Join the encoded columns with the original dataframe
df_encoded = pd.concat([df_encoded, encoded_nominal_df], axis=1)

# Drop the original nominal columns
df_encoded.drop(nominal_cols, axis=1, inplace=True)

# Display the shape after one-hot encoding
print(f"DataFrame shape after one-hot encoding: {df_encoded.shape}")
```

### Cyclical Encoding

```python
# Months and days of week have cyclical nature
import numpy as np

# Month encoding
month_order = {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12}
df_encoded['month_num'] = df_encoded['month'].map(month_order)

# Encode month cyclically
df_encoded['month_sin'] = np.sin(2 * np.pi * df_encoded['month_num'] / 12)
df_encoded['month_cos'] = np.cos(2 * np.pi * df_encoded['month_num'] / 12)

# Day of week encoding
day_order = {'mon':1, 'tue':2, 'wed':3, 'thu':4, 'fri':5, 'sat':6, 'sun':7}
df_encoded['day_num'] = df_encoded['day_of_week'].map(day_order)

# Encode day of week cyclically
df_encoded['day_sin'] = np.sin(2 * np.pi * df_encoded['day_num'] / 7)
df_encoded['day_cos'] = np.cos(2 * np.pi * df_encoded['day_num'] / 7)

# Drop the original month and day_of_week columns and the intermediate numeric columns
df_encoded.drop(['month', 'day_of_week', 'month_num', 'day_num'], axis=1, inplace=True)

# Show the first few rows after cyclical encoding
df_encoded[['month_sin', 'month_cos', 'day_sin', 'day_cos']].head()
```

### Feature Engineering

```python
# Feature 1: Campaign Efficiency Ratio (captures how many calls were needed in relation to previous campaigns)
df_encoded['campaign_previous_ratio'] = df_encoded['campaign'] / (df_encoded['previous'] + 1)  # Adding 1 to avoid division by zero

# Feature 2: Age groups (categorizing age into meaningful segments)
bins = [17, 30, 40, 50, 60, 100]  
labels = ['18-30', '31-40', '41-50', '51-60', '60+']
df_encoded['age_group'] = pd.cut(df_encoded['age'], bins=bins, labels=labels)

# One-hot encode the age groups
age_dummies = pd.get_dummies(df_encoded['age_group'], prefix='age_group', drop_first=True)
df_encoded = pd.concat([df_encoded, age_dummies], axis=1)
df_encoded.drop('age_group', axis=1, inplace=True)

# Feature 3: Economic Health Index 
# Normalize the economic indicators first
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
economic_features = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
df_encoded[economic_features] = scaler.fit_transform(df_encoded[economic_features])

# Create a positive economic sentiment score 
df_encoded['economic_sentiment'] = (df_encoded['emp.var.rate'] + df_encoded['cons.conf.idx'] + 
                                   df_encoded['nr.employed'] - abs(df_encoded['cons.price.idx']))

# Feature 4: Contact Attempt Features
df_encoded['has_previous_contact'] = np.where(df_encoded['pdays'] == 999, 0, 1)

# Show the first few rows of the new features
new_features = ['campaign_previous_ratio', 'economic_sentiment', 'has_previous_contact']
df_encoded[new_features].head()
```

```python
# Separate features and target variable
X = df_preprocessed.drop('y', axis=1)
y = df_preprocessed['y']

# Import necessary libraries
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN, SMOTETomek
from collections import Counter

# Split the data with 20% for testing using stratified sampling to maintain class distribution
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

# Check class distribution in training and test sets
print("Training set class distribution:")
print(Counter(y_train))
print("\nTest set class distribution:")
print(Counter(y_test))
```

```python
# Apply SMOTE to handle class imbalance in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check class distribution after SMOTE
print("Class distribution after SMOTE:")
print(Counter(y_train_smote))

# Apply Random Under-Sampling for comparison
rus = RandomUnderSampler(random_state=42)
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

# Check class distribution after Random Under-Sampling
print("\nClass distribution after Random Under-Sampling:")
print(Counter(y_train_rus))

# Apply combination method: SMOTE + ENN (Edited Nearest Neighbors)
smoteenn = SMOTEENN(random_state=42)
X_train_smoteenn, y_train_smoteenn = smoteenn.fit_resample(X_train, y_train)

# Check class distribution after SMOTEENN
print("\nClass distribution after SMOTE + ENN:")
print(Counter(y_train_smoteenn))
```

# Part 4: Model Training and Evaluation

Let's train multiple models on our oversampled data and select the best one based on performance metrics.

```python
# Load the saved train-test splits
with open('train_test_splits.pkl', 'rb') as f:
    train_test_splits = pickle.load(f)
    
# We'll use the SMOTE version for our model training
X_train = train_test_splits['smote']['X_train']
X_test = train_test_splits['smote']['X_test']
y_train = train_test_splits['smote']['y_train']
y_test = train_test_splits['smote']['y_test']

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
# Import models and evaluation metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# Create a list of models to train
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'MLP Classifier': MLPClassifier(max_iter=1000, random_state=42)
}
```

```python
# Create a function to evaluate models
def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    # Get unique classes
    classes = np.unique(np.concatenate([y_train, y_test]))
    pos_label = classes[-1]  # Use the highest value as positive class
    
    # Calculate metrics with explicit pos_label
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, pos_label=pos_label)
    rec = recall_score(y_test, y_pred, pos_label=pos_label)
    f1 = f1_score(y_test, y_pred, pos_label=pos_label)
    roc_auc = roc_auc_score(y_test, y_pred_prob) if y_pred_prob is not None else None
    
    return {
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1_score': f1,
        'roc_auc': roc_auc,
        'model': model,
        'predictions': y_pred,
        'prob_predictions': y_pred_prob
    }
```

```python
# Train and evaluate models
results = {}

for name, model in models.items():
    print(f"Training {name}...")
    results[name] = evaluate_model(model, X_train, X_test, y_train, y_test)
    print(f"Completed training {name}.\n")
```

```python
# Create a DataFrame to compare model performance
results_df = pd.DataFrame({
    'Model': list(results.keys()),
    'Accuracy': [results[model]['accuracy'] for model in results],
    'Precision': [results[model]['precision'] for model in results],
    'Recall': [results[model]['recall'] for model in results],
    'F1 Score': [results[model]['f1_score'] for model in results],
    'ROC AUC': [results[model]['roc_auc'] for model in results]
})

# Sort by F1 Score (balanced metric)
results_df_sorted = results_df.sort_values('F1 Score', ascending=False).reset_index(drop=True)
results_df_sorted
```

```python
# Visualize the model comparison
plt.figure(figsize=(15, 10))

# Plot different metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC']
for i, metric in enumerate(metrics):
    plt.subplot(2, 3, i+1)
    sns.barplot(x='Model', y=metric, data=results_df_sorted)
    plt.title(f'Model Comparison - {metric}')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 1.0)
    
plt.tight_layout()
plt.suptitle('Model Performance Comparison', fontsize=16, y=1.05)
plt.show()
```

```python
# Get the best model based on F1 score
best_model_name = results_df_sorted['Model'].iloc[0]
best_model = results[best_model_name]['model']

print(f"Best performing model: {best_model_name}")
print("\nClassification Report:")
print(classification_report(y_test, results[best_model_name]['predictions']))

# Plot confusion matrix for the best model
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, results[best_model_name]['predictions'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title(f'Confusion Matrix - {best_model_name}')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks([0.5, 1.5], ['No', 'Yes'])
plt.yticks([0.5, 1.5], ['No', 'Yes'])
plt.show()
```
